```python
import pandas as pd
```

```python
df = pd.read_csv('Zomato_Dataset.csv')
df.head()
```

```python
df['Restaurant_distance'] = ((df['Restaurant_latitude'] - df['Delivery_location_latitude'])**2 + (df['Restaurant_longitude'] - df['Delivery_location_longitude'])**2)**0.5
```

```python
df.head()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')

```

```python
def fraction_to_time(fraction):
    """Convert fraction of a day (0–1) into HH:MM format."""
    if pd.isna(fraction):
        return None
    total_minutes = int(round(float(fraction) * 24 * 60))
    hours = total_minutes // 60
    minutes = total_minutes % 60
    return f"{hours:02d}:{minutes:02d}"


# Check the datatype of the column
if pd.api.types.is_numeric_dtype(df['Time_Orderd']):
    # Column is numeric — treat as fraction of a day
    df['Time_Orderd'] = df['Time_Orderd'].apply(fraction_to_time)


df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'], errors='coerce', format='%H:%M')

# Check the datatype of the column
if pd.api.types.is_numeric_dtype(df['Time_Order_picked']):
    # Column is numeric — treat as fraction of a day
    df['Time_Order_picked'] = df['TTime_Order_picked'].apply(fraction_to_time)


df['Time_Order_picked'] = pd.to_datetime(df['Time_Order_picked'], errors='coerce', format='%H:%M')

df['Time_Order_picked']
```
