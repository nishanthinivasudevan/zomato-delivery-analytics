```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')

```

```python
def fraction_to_time(fraction):
    """Convert fraction of a day (0–1) into HH:MM format."""
    if pd.isna(fraction):
        return None
    total_minutes = int(round(float(fraction) * 24 * 60))
    hours = total_minutes // 60
    minutes = total_minutes % 60
    return f"{hours:02d}:{minutes:02d}"


# Check the datatype of the column
if pd.api.types.is_numeric_dtype(df['Time_Orderd']):
    # Column is numeric — treat as fraction of a day
    df['Time_Orderd'] = df['Time_Orderd'].apply(fraction_to_time)


df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'], errors='coerce', format='%H:%M')

# Check the datatype of the column
if pd.api.types.is_numeric_dtype(df['Time_Order_picked']):
    # Column is numeric — treat as fraction of a day
    df['Time_Order_picked'] = df['TTime_Order_picked'].apply(fraction_to_time)


df['Time_Order_picked'] = pd.to_datetime(df['Time_Order_picked'], errors='coerce', format='%H:%M')

df['Time_Order_picked']
```

```python
# Mode based imputation
Weather_mode = df['Weather_conditions'].mode()[0]
df.fillna({'Weather_conditions': Weather_mode}, inplace=True)

Road_traffic_mode = df['Road_traffic_density'].mode()[0]
df.fillna({'Road_traffic_density': Road_traffic_mode}, inplace=True)

Festival_mode = df['Festival'].mode()[0]
df.fillna({'Festival': Festival_mode}, inplace=True)

City_mode = df['City'].mode()[0]
df.fillna({'City': City_mode}, inplace=True)
```

```python
# Mean based imputation
Rating_mean = df['Delivery_person_Ratings'].mean()
df.fillna({'Delivery_person_Ratings': Rating_mean}, inplace=True)
df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].round(decimals=1)

age_mean = int(df['Delivery_person_Age'].mean())
df.fillna({'Delivery_person_Age': age_mean}, inplace=True)

del_mean = int(df['multiple_deliveries'].mean())
df.fillna({'multiple_deliveries': del_mean}, inplace=True)
```

```python
max_rating = df['Delivery_person_Ratings'].unique()
max_rating
# Some ratings are given as 6 which is incorrect as max rating is 5

df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].replace(6, 5)
```
